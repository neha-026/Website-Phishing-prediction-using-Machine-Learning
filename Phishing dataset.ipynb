#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
ac=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
```

```python
print("Accuracy:",ac)
print("Confusion Matrix:",cm)
print("Classification Report:",cr)
```

```python
train=logreg.score(xtrain,ytrain)
test=logreg.score(xtest,ytest)

print("Training Performance:",train)
print("Testing Performance:",test)
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

from sklearn.metrics import accuracy_score, classification_report
logreg=LogisticRegression()
svm=SVC()
dt=DecisionTreeClassifier()
```

```python
def mod(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    
    cr=classification_report(ytest,ypred)
    print(f"{cr}")
```

```python
mod(logreg)
```

```python
mod(dt) #here DT is giving us the highest accuracy
```

```python
mod(svm) # we are not performing SVM because it has decreased the accuracy rate
```

```python
dt.score(xtrain,ytrain)
```

```python
dt.score(xtest,ytest)
```

```python
# As the DT model is overfitted, we will use the hypertuning techniques 
```

```python
for i in range(1,31):
    dt1=DecisionTreeClassifier(max_depth=i)         
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f"max_depth={i} accuracy={ac}")
```

```python
dt2=DecisionTreeClassifier(max_depth=4)
mod(dt2)
```

```python
dt2.score(xtrain,ytrain)
```

```python
dt2.score(xtest,ytest)                                           #using max_depth we are getting 92% accuracy at max_depth=4
```

```python
for i in range(80,110):
    dt3=DecisionTreeClassifier(min_samples_leaf=i)
    dt3.fit(xtrain,ytrain)
    ypred=dt3.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f"max_depth={i}  accuracy_score={ac}")
```

```python
dt4=DecisionTreeClassifier(min_samples_leaf=105)
mod(dt4)
```

```python
dt4.score(xtrain,ytrain)                                           #min_sample_leaf=105 is giving us the 92% accuracy of model
```

```python
dt4.score(xtest,ytest)
```

```python
for i in range (50,100):
    dt5=DecisionTreeClassifier(min_samples_split=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    
    ac=accuracy_score(ytest,ypred)
    print(f"max_depth={i} accuracy={ac}")
```

```python
dt6=DecisionTreeClassifier(min_samples_split=221)
mod(dt6)
```

```python
dt6.score(xtrain,ytrain)                                      # min_sample_split=221 gives 92% accuracy
```

```python
dt6.score(xtest,ytest)
```

### Bagging

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
```

```python
bg=BaggingClassifier(DecisionTreeClassifier())                               #95 percent accuracy using DT
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)
cr=classification_report(ytest,ypred)
print(cr)
```

```python
bg=BaggingClassifier(LogisticRegression())                                  # logistic Regression is giving 80 percent accuracy
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)
cr=classification_report(ytest,ypred)
print(cr)
```

```python
rfc=RandomForestClassifier()
rfc.fit(xtrain,ytrain)
ypred=rfc.predict(xtest)
cr=classification_report(ytest,ypred)                         #it is giving the accuracy of 96%
print(cr)
```
